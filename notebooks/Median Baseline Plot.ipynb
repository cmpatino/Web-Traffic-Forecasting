```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

plt.style.use('seaborn-poster')
register_matplotlib_converters()
```

```python
train_df = pd.read_csv('../data/input/train_2.csv')
train_df = train_df.set_index('Page')
```

```python
sample_series = train_df.iloc[9:10]
sample_series = sample_series.transpose()
sample_series.index = pd.DatetimeIndex(sample_series.index)
sample_series.columns = ['Visits']
sample_series.head()
```

```python
pred = [np.median(sample_series.values[-90:])]*62
x_pred = pd.date_range('09/10/2017', '11/10/2017')
```

```python
ax = plt.axes()
plt.plot(sample_series.index[-200:], sample_series.Visits[-200:], label='Historic')
plt.plot(x_pred, pred, label='Prediction')
ax.axvspan(sample_series.index[-90], sample_series.index[-1], alpha=0.5, color='orange')
plt.legend(fontsize=26)
plt.ylabel('Visits', fontsize=30)
plt.xlabel('Date', fontsize=30)
```
